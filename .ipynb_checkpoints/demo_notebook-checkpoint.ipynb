```python

# Silent Meeting Observer - Demo Notebook

from app.classifier import batch_classify
from app.risk import score_risk
from app.render import render_markdown
import json
```

```python
with open("sample_data/meeting.json") as f:
    meeting = json.load(f)
```

```python
# Run classifier on utterances
classified = batch_classify(meeting["utterances"])

print("Classified utterances:")
for item in classified:
    print(f"[{item['type']}] {item['speaker']}: {item['text']}")
```

```python
# Apply risk scoring
scored = [score_risk(it) for it in classified]

print("Risks identified:")
for r in scored:
    if r["type"] == "Risk":
        print(f"- {r['text']} (p={r['probability']}, severity={r['severity']})")
```

```python
# Render final smart summary in Markdown format
summary = render_markdown(meeting, scored)
print(summary)
```
